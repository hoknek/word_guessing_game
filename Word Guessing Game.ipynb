```python
import random
import time

def guessing_game(word):

    error = 0
    #guess_left = (7 - int(error)) #why does this not work?
    given_word = list(word)
    made_word = "_" * int(len(word))
    string_made = list(made_word)
    letters_tried = []
    repeats = []
    
    print("-" * 40)
    print("Welcome to the letter guessing game!")
    print("Selecting random word...")
    print("-" * 40)
    time.sleep(1.5)

    done = False
    while not done:
        guess = input("\nGuess a letter: ")
        if guess.isalpha() == False or len(guess) > 1:
            print("!!! Invalid entry. Please type in a letter. !!!")
            print("    - Letters tried - " + str(letters_tried))
            print("You have " + str(7 - error) + " guesses left.") 
        elif guess not in word:
            if guess in letters_tried:
                print("You tried that letter already. Try another letter!")
                print("    - Letters tried - " + str(letters_tried))
                print("You have " + str(7 - error) + " guesses left.") 
            elif guess not in letters_tried:
                error += 1
                if error >= 7: 
                    print('\n"' + guess + '" is not in the word.')
                    print("    - Letters tried - " + str(letters_tried))
                    print("You have no more guesses...")
                    print("=" * 40)
                    print(" ".join(string_made).upper() + ('      (The word was "' + word + '".)'))
                    print("=" * 40)
                    print("Game over...")
                    print("--  :(  --" * 4)
                    done = True
                else: 
                    if guess not in letters_tried:
                        letters_tried.append(guess)
                    print('\n"' + guess + '" is not in the word. Try again!')
                    print("    - Letters tried - " + str(letters_tried))
                    print("You have " + str(7 - error) + " guesses left.")               
        elif guess in word:
            if guess in letters_tried:
                print("You tried that letter already. Try another letter!")
                print("    - Letters tried - " + str(letters_tried))
                print("You have " + str(7 - error) + " guesses left.") 
            else:
                for index, i in enumerate(word):
                    if i == guess:
                        letters_tried.append(guess)
                        if i not in repeats:
                            string_made[index] = i
                            repeats.append(i)
                            if given_word == string_made:
                                print("\nNice job!")
                                print("    - Letters tried - " + str(letters_tried))
                                print("You got it!")
                                print("=" * 40)
                                print(" ".join(string_made).upper())
                                print("=" * 40)
                                print("You win!")
                                print("--  :)  --" * 4) 
                                done = True 
                            else: 
                                print("\nNice job! Keep going!")
                                print("    - Letters tried - " + str(letters_tried))
                                print("You have " + str(7 - error) + " guesses left.")                   
                        else:
                            string_made[index] = i
        if error < 7: 
            if given_word != string_made:
                print("=" * 40)
                print(" ".join(string_made).upper())
                print("=" * 40)
        
        
words = ['apple','bee', 'cat','dinasaur','eggs','ferry','gaters','hindges','iris','jackhammer','kitkat']
word = random.choice(words)
guessing_game(word)
```
